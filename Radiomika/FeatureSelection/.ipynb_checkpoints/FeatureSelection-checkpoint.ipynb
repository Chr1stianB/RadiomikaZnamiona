```python
import pandas as pd
import os
from imutils import paths
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso
import radiomics
from radiomics import featureextractor  # This module is used for interaction with pyradiomics
from PIL import Image
import SimpleITK as sitk
import cv2
from collections import Counter
import random
import itertools
```

```python
DATASET_PATH = "../Datasets/ham1000-segmentation-and-classification"
CSV_FILE_PATH = os.path.join(DATASET_PATH, "GroundTruth.csv")
# define the path to the images and masks dataset
IMAGE_DATASET_PATH = os.path.join(DATASET_PATH, "images/")
MASK_DATASET_PATH = os.path.join(DATASET_PATH, "masks/")

# Type of tumor
melanoma = []
nevus = []
basal_cell_carcinoma = []
actinic_keratoses_and_intraepithelial_carcinoma = []
benign_keratosis_like_lesions = []
dermatofibroma  = []
vascular_lesion = []

lesion_dict = {
    "MEL": 0, 
    "NV": 1,
    "BCC": 2,
    "AKIEC": 3,
    "BKL": 4,
    "DF": 5,
    "VASC": 6
}

new_lesion_dict = {
    0: [], 
    1: [],
    2: [],
    3: [],
    4: [],
    5: [],
    6: []
}


# define the test split
TEST_SPLIT = 0.15
params = 'Params.yaml'
```

```python
imagePaths = sorted(list(paths.list_images(IMAGE_DATASET_PATH)))
maskPaths = sorted(list(paths.list_images(MASK_DATASET_PATH)))

split = train_test_split(imagePaths, maskPaths,
    test_size=TEST_SPLIT, random_state=42)

# unpack the data split
(trainImages, testImages) = split[:2]
(trainMasks, testMasks) = split[2:]
  
```

```python
extractor = featureextractor.RadiomicsFeatureExtractor(params)
results = []
features = {}

for i in range(len(imagePaths)):
    path_original = imagePaths[i]
    path_label = maskPaths[i]
    original = cv2.imread(path_original)
    original = cv2.cvtColor(original, cv2.COLOR_RGB2GRAY)
    #original = original.squeeze()
    original = Image.fromarray(original, "L")
    original.save("original.png")
    #original = image.astype("float32") / 255.0
    #original = sitk.ReadImage(path_original)
    #original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
    #label = cv2.cvtColor(label, cv2.COLOR_BGR2GRAY)

    #label = cv2.imread(path_label, cv2.IMREAD_GRAYSCALE)
    label = cv2.imread(path_label, 0)
    label = Image.fromarray(label, "L")
    label.save("label.png")
    #print(label.shape)
    #label = sitk.ReadImage(path_label)
    result = extractor.execute("original.png", path_label)
    results.append(result)
    
feature_names = list(sorted(filter ( lambda k: k.startswith("original_"), results[0] )))
```

### Wektor wartości wyjściowych
#### Wszystkie 10013 znamion z numerami od 0-6, oznaczającymi kategorie znamienia skórnego

```python
df = pd.read_csv(CSV_FILE_PATH)
categories = []
for index, row in df.iterrows():
    if row.eq(1).any():
        categories.append(lesion_dict[row[row.eq(1)].index[0]])
```

### Wartości cech radiomicznych dla wszystkich próbek 

```python
samples = np.zeros((len(imagePaths),len(feature_names)))
for case_id in range(0 ,len(imagePaths)):
    a = np.array([])
    for feature_name in feature_names:
        a = np.append(a, results[case_id][feature_name])
    samples[case_id,:] = a
    
# May have NaNs
samples = np.nan_to_num(samples)
```

```python
for i in range(len(categories)):
    new_lesion_dict[categories[i]].append(samples[i])
        
```

```python
X =[]
y = []
for category in new_lesion_dict:
    for i in range(100):
        X.append(new_lesion_dict[category][i])
        y.append(category)
        
temp = list(zip(X, y))
random.shuffle(temp)
res1, res2 = zip(*temp)
# res1 and res2 come out as tuples, and so must be converted to lists.
X, y = list(res1), list(res2)
```

```python
pipeline = Pipeline([
                     ('scaler',StandardScaler()),
                     ('model',Lasso())
])
```

```python
search = GridSearchCV(pipeline,
                      {'model__alpha':np.arange(0.1,10,0.1)},
                      cv = 5, scoring="neg_mean_squared_error",verbose=3
                      )
```

```python
search.fit(X,y)
```

```python
search.best_params_
```

```python
coefficients = search.best_estimator_.named_steps['model'].coef_
```

```python
importance = np.abs(coefficients)
```
